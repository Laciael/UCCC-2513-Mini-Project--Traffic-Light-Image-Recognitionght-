```python
#image segmented from orginal to grayscale and binary

#Scale image as grayscale to reduce the amount of RGB pixelation required for segregation
import cv2 as cv
import numpy as np

img1 = cv.imread('Cars.jpg') #default photo
img2 = cv.imread('Cars.jpg', 0)    # synchronize dimension

#show photo dimension size
print(img1.shape)
print(img2.shape)

#resize photo
img1 = cv.resize(img1, (0,0), None, 0.2, 0.2)
img2 = cv.resize(img2, (0,0), None, 0.2, 0.2)

#convert to grayscale
img2 = cv.cvtColor(img2, cv.COLOR_GRAY2BGR)

#Finding edges on images
img3=cv.Canny(img2, 30, 200)

#use a copy of your image, e.g. - edged.copy(), since finding contours alter the image
#we have to add _, before the contours as an empty argument due to upgrade of the OpenCV version
_, contours=cv.findContours(img3,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_NONE)

#print contour
print(contours)
print('Numbers of contours found=' + str(len(contours)))

#output
cv.imshow('Orignal_Cars', img1)
cv.imshow('Grayscale_Cars', img2)
cv.imshow('Binary_Cars', img3)

cv.imwrite('Orignal_Cars.jpg', img1)
cv.imwrite('Grayscale_Cars.jpg', img2)
cv.imwrite('Binary_Cars.jpg', img3)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#test prototype

#Scale image as grayscale to reduce the amount of RGB pixelation required for segregation
import cv2 as cv
import numpy as np

car_cascade = cv.CascadeClassifier('cars.xml')  #Path to cars.xml

img1 = cv.imread('R.jpg') #default photo
img2 = cv.imread('R.jpg', 0)    # synchronize dimension

#show photo dimension size
print(img1.shape)
print(img2.shape)

#resize photo
img1 = cv.resize(img1, (0,0), None, 0.5, 0.5)
img2 = cv.resize(img2, (0,0), None, 0.5, 0.5)

#convert to grayscale
img2 = cv.cvtColor(img2, cv.COLOR_GRAY2BGR)

#Finding edges on images
img3=cv.Canny(img2, 30, 200)

#use a copy of your image, e.g. - edged.copy(), since finding contours alter the image
#we have to add _, before the contours as an empty argument due to upgrade of the OpenCV version
_, contours=cv.findContours(img3,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_NONE)

#print contour
print(contours)
print('Numbers of contours found=' + str(len(contours)))

cars=car_cascade.detectMultiScale(img3,1.8,2)

#Drawing rectangles on detected cars
for (x,y,w,h) in cars:
    cv.rectangle(img3,(x,y),(x+w,y+h),(225,0,0),2)

#output
cv.imshow('R', img1)
cv.imshow('Grayscale_R', img2)
cv.imshow('SurprisedVR', img3)
cv.imshow('cars', cars)

cv.waitKey(0)
cv.destroyAllWindows()
```
